```python
import gradio as gr           #UI Library  
from transformers import pipeline  #Transformers library
```

```python
translation_pipeline = pipeline('translation_en_to_fr')
```

```python
text = "Hello,My Name is Aseel, And I love Artificial intelligence"
translated_text = translation_pipeline(text)[0]['translation_text']
```

```python
#import gradio as gr
#from transformers import pipeline

# Function to perform translation using transformers pipeline
def translate_transformers(text):
    translation_pipeline = pipeline('translation_en_to_fr')  # Example translation from English to French
    translated_text = translation_pipeline(text)[0]['translation_text']
    return translated_text

# Create Gradio interface
interface = gr.Interface(
    fn=translate_transformers,
    inputs=gr.Textbox(lines=2, placeholder='Text to translate'),
    outputs=gr.Textbox(label='Translated Text')
)

```

```python
# Launch  interface with sharing enabled
interface.launch(share=True)
```
